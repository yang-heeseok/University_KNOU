```python
#반복문
# for
for n in range(1, 6):          
    print(n)

# while
n = 1
while n <= 5:
    print(n)
    n = n+1
```

```python
# while - continue
n = 1
while n <= 5:
    if n == 4:
        n = n+1
        continue
    print(n)
    n = n+1
```

```python
# while - break
n = 1
while n <= 5:
    if n == 4:        
        n = n+1
        break
    print(n)
    n = n+1
```

```python
# 조건문 & 함수 작성
def mysign(x):
    if x > 0:
        print(f"{x} : 양수")
    elif x < 0:
        print(f"{x} : 음수")
    else:
        print("0입니다")
        
mysign(1)
mysign(-3)
mysign(0)
```

```python
# 조건문 & 함수 작성
def mywage(h):    
    r1= 10000    
    r2 = r1 * 1.5    
    if h <= 40:        
        wage = h * r1    
    else:       
        wage = 40 * r1 + (h - 40) * r2
    return wage



print(mywage(35),"원")
print(mywage(45),"원")
```
